```python
import pandas as pd
import os

import numpy as np
```

```python
file1="NewComboJanFebMar2.csv"
file2="NewComboJunJulAug.csv"
a=[file1, file2]
```

```python
frames=[]

for y in range (0,(len(a))):
        print ("processing file: " +str(y+1))
        dataAll=pd.DataFrame(pd.read_csv(a[y]))
        frames.append("dataAll" +str(y))
        frames[y]=dataAll
```
